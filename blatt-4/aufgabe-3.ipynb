```python
import numpy as np
import matplotlib.pyplot as plt
```

### SVM-Klassifikation

```python
x1 = np.array([[1], [1]])
x2 = np.array([[3], [3]])
x3 = np.array([[1], [3]])
x4 = np.array([[3], [1]])
x5= np.array([[2], [2.5]])
x6 = np.array([[3], [2.5]])
x7 = np.array([[4], [3]])

class1 = [x2, x3, x5]
class2 = [x1, x4, x6, x7]

for vector in class1:
    plt.scatter(vector[0], vector[1], color='red', label='Klasse 1')

for vector in class2:
    plt.scatter(vector[0], vector[1], color='blue', label='Klasse 2')

plt.plot([0, 3.5], [1, 3.0], color='green', linestyle='--', label='Trennfunktion')

plt.xlabel('X-Komponente')
plt.ylabel('Y-Komponente')

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='lower right')
plt.ylim(0, 3.2)
plt.title('Plot der Vektoren nach Klassen')
plt.show()
```
